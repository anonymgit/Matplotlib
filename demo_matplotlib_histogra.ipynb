# plotting histograms

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,40,(100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
#plotting histogram
plt.hist(ml_student_age)
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")
plt.show()
```

```python
#bins-showing in bar form
#rwidth -gives gap between bars(percent)
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8)
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")

plt.show()
```

```python
#histogram types
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='step')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")

plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,histtype='bar')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")

plt.show()
```

```python
#orientation
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,orientation='horizontal')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")

plt.show()
```

```python
#color-change color of bar
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,color='red')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")

plt.show()
```

```python
#label-label histograms
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.8,label='ML students')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,color='r',rwidth=0.8,label='ML students')
plt.hist(py_student_age,bins,color='m',rwidth=0.8,label='py students')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_student_age,py_student_age],bins,color=['r','y'],rwidth=0.8,label=['ML students','py_students'])
#plt.hist(py_student_age,bins,color='m',rwidth=0.8,label='py students')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([ml_student_age,py_student_age],bins,color=['r','y'],rwidth=0.8,label=['ML students','py_students'])
#plt.hist(py_student_age,bins,color='m',rwidth=0.8,label='py students')
plt.title("ML students age histograms")
plt.xlabel("students age category")
plt.ylabel("no. students age")
plt.legend()
plt.show()
```
