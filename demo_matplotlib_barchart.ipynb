# Bar chart()

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#dataset of indian Artificial intelligence
#classes=["python","R","AI","ML","DS"]
classes=["python","R","AI","ML","DS"]
class1_students=[10,10,20,25,10]
class2_students=[40,5,20,20,10]
class3_students=[35,5,30,15,15]
```

```python
#plt.bar(on-x axis,on-y axis)   vertical bar chart
plt.bar(classes,class1_students)
```

```python
#plt.bar(on-x axis,on-y axis)   horizontal bar chart
plt.barh(classes,class1_students)
```

```python
#use some parameters of bar chart
plt.bar(classes,class1_students,width=0.2)  #change bar width to 20%
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color='y')
```

```python
#edge color change
plt.bar(classes,class1_students,width=0.2,align="edge",color='y',edgecolor='m')
```

```python
#line width of edge 
plt.bar(classes,class1_students,width=0.2,align="edge",color='y',edgecolor='m',linewidth=5)
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color='y',edgecolor='m',linestyle='--')
```

```python
plt.bar(classes,class1_students,width=0.2,align="edge",color='y',edgecolor='m',label="class1 students")
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.6,align="edge",color='y',edgecolor='m',label="class1 students")
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.6,align="edge",color='y',edgecolor='m',label="class1 students")
plt.title("bar char Demo",fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel("No of student",fontsize=20)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1_students,width=0.6,align="edge",color='y',edgecolor='m')
plt.bar(classes,class2_students,width=0.6,align="edge",color='b',edgecolor='m')
plt.bar(classes,class3_students,width=0.6,align="edge",color='g',edgecolor='m')
plt.title("bar char Demo",fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel("No of student",fontsize=20)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
width=0.2
plt.bar(classes_index,class1_students,width=0.2,align="edge",color='y',edgecolor='m')
plt.bar(classes_index+width,class2_students,align="edge",color='b',edgecolor='m')
plt.bar(classes_index+width+width,class3_students,align="edge",color='g',edgecolor='m')
plt.title("bar char Demo",fontsize=20)
plt.xlabel("classes",fontsize=20)
plt.ylabel("No of student",fontsize=20)
plt.show()
```
